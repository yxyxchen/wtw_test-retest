```python
# inputs
sess = 1
hdrdatafile = os.path.join("data", 'hdrdata_sess%d.csv'%sess)
trialdatafiles = glob.glob(os.path.join("data", "task*.csv"))
selfreportfile = os.path.join("data", 'selfreport_sess%d.csv'%sess)
```

```python
# process selfreport data
selfreport = pd.read_csv(selfreportfile)
selfdata = pd.DataFrame()
for i, row in selfreport.iterrows():
    # calculate k
    try:
        k,logk,se_logk  = calc_k(row[4:27+4], False)
    except:
        k = None,
        logk = None,
        se_logk = None

    NU, PU, PM, PS, SS, UPPS  = score_upps(row[61:120], False)
    
    # calculate UPPS
    selfdata = pd.concat([selfdata, pd.DataFrame({
        "k":k,
        "logk": k,
        "se_logk": se_logk,
        "NU": NU,
        "PU": PU,
        "PM": PM,
        "PS": PS,
        "SS": SS,
        "UPPS": UPPS
        }, index = [i])])
```

```python
# load taskdata
hdrdata   = pd.read_csv(hdrdatafile)
nsub = len(trialdatafiles)
trialdata  = {}
for i in range(nsub):
    thisTrialData = pd.read_csv(trialdatafiles[i])
    
    # add totalTrialIdx
    thisTrialData['totalTrialIdx'] = thisTrialData.trialIdx
    ntrial_firstblock = int(thisTrialData.trialIdx[thisTrialData.condition == "LP"].max())
    thisTrialData['totalTrialIdx'].iloc[ntrial_firstblock:] = thisTrialData['totalTrialIdx'].iloc[ntrial_firstblock:] +  ntrial_firstblock 
    
    # fill in 
    trialdata[(hdrdata.id[i], hdrdata.sess[i])] =  thisTrialData
```

```python
# I want an individual loop here.
# response time check
fig, ax = plt.subplots(1, 10)
fig.set_figwidth(15)
for i in range(nsub):
    thisTrialData = trialdata[(hdrdata.id[i], hdrdata.sess[i])]
    
    n, bins, patches = ax[i].hist(x=thisTrialData.trialStartTime - thisTrialData.trialReadyTime , bins='auto', color='#0504aa',
                                alpha=0.7, rwidth=0.85)
#     n, bins, patches = ax[i].hist(x=thisTrialData.RT, bins='auto', color='#0504aa',
#                                 alpha=0.7, rwidth=0.85)
    ax[i].set_xlabel('Ready RT (s)')
    #ax[i].set_ylabel('Frequency')
    ax[i].set_title(hdrdata.id[i])
    # input("Press Enter to continue...")
    
# there might be an overall energy level covariate with time or condition changes
# whether these two measures agree with each other
# In general I am not too worried about ready though ...
fig, ax = plt.subplots(10, 1)
fig.set_figheight(15)

for i in range(nsub):
    thisTrialData = trialdata[(hdrdata.id[i], hdrdata.sess[i])]
    
    #ax[i].plot(thisTrialData.totalTrialIdx, thisTrialData.RT)
    ax[i].plot(thisTrialData.totalTrialIdx,  (thisTrialData.trialStartTime - thisTrialData.trialReadyTime))
    ax[i].set_title(hdrdata.id[i])
```
